### Support Vector Classifier (SVC) Implementation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples= 1000, n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x= pd.DataFrame(X)[0], y= pd.DataFrame(X)[1], hue= y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear') # Linear Kernel = making a straight line or best fit line
```

```python
svc.fit(X_train,y_train)
```

```python
# Coefficients of the best fit line it shoes the slope of the line and intercept of the line. 
#Two features so two coefficients

svc.coef_ 
```

```python
## prediction 
y_pred= svc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

### Hyperparameter Tuning with SVC

```python
from sklearn.model_selection import GridSearchCV

## defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
                'kernel': ['linear']
              }
```

```python
grid = GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv = 5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python

grid.best_params_
```

```python
grid.best_estimator_
```

```python
## Predcition 
grid_pred = grid.predict(X_test)
print(classification_report(y_test,grid_pred))
print(confusion_matrix(y_test,grid_pred))
print(accuracy_score(y_test,grid_pred))
```
