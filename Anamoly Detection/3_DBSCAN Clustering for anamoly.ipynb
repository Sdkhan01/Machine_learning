```python
## DBSCAN Clustering for anamoly detection

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
from sklearn.datasets import make_circles
%matplotlib inline
```

```python
# Create data

X,y = make_circles(n_samples = 750, factor= 0.3, noise = 0.1)
```

```python
X.shape
```

```python
# Plot data

plt.scatter(X[:,0],X[:,1],c = y)
```

```python
## makiing the model using DBSCAN

dbscan = DBSCAN(eps= 0.1)
```

```python
dbscan.fit_predict(X)
```

```python
dbscan.labels_
# it will give the labels of the data points
# -1 is the outlier
#labels are the cluster number
#labels in DBSCAN are the cluster number it helps to identify the cluster number
```

```python
plt.scatter(X[:,0],X[:,1],c = dbscan.labels_)
```
