## Local Outlier Factror Anamoly Detection

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_circles
from sklearn.datasets import make_moons
```

```python
# Generate some data
X,y = make_moons(n_samples=750, noise=0.1)
```

```python
X.shape
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
# Fit the model

lof = LocalOutlierFactor(n_neighbors=2, contamination= 'auto')
lof
```

```python
#fit the model
fit = lof.fit(X)
fit
```

```python
#fit the model
pred = lof.fit_predict(X)
```

```python
# Plot the data
plt.scatter(X[:,0],X[:,1], c=pred)
```

```python
plt.scatter(X[:,0],X[:,1],c = fit.negative_outlier_factor_)
```

```python
lof.negative_outlier_factor_
```

```python
# Generate some data
X,y = make_circles(n_samples=750, noise=0.1)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
# Fit the model

lof = LocalOutlierFactor(n_neighbors=2, contamination= 'auto')
```

```python
#fit the model
fit = lof.fit(X)
```

```python
#fit the model
pred = lof.fit_predict(X)
```

```python
# Plot the data
plt.scatter(X[:,0],X[:,1], c=pred)
```

```python
## Conclusion from the above two plots

## The first plot is a moon shaped data and the second plot is a circle shaped data. The local outlier factor is able to detect the outliers in the moon shaped data but not in the circle shaped data. 
## and the reason is that the local outlier factor is not able to detect the outliers in the circle shaped data because the local outlier factor is not able to detect the outliers in the data that is not dense.
```
