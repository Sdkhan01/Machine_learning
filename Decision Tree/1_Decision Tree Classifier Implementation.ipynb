## Decision Tree Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()   
```

```python
dataset.target #the target column consist of the target values which are the species of the iris flower
```

```python
## independent and dependent features

X = df.iloc[:,:-1] ## independent features,#taking all the columns except the last one
y = dataset.target ## dependent features #taking the last column from the dataset which is the target column

```

```python
X,y
```

```python
## Train test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```


### We dont need to do feature scaling in deciosion Trees
### means,dont need  to do standardization or normalization
 

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
## Post Pruning,This is method to avoid overfitting
tree_classifier = DecisionTreeClassifier()

```

```python
tree_classifier.fit(X_train,y_train)
```

```python
X_train.head()
```

```python
from sklearn import tree
plt.figure(figsize = (50,50))
tree.plot_tree(tree_classifier,filled = True)
```

```python
X_train.head()
```

```python
## Post Pruing with max depth

tree_classifier = DecisionTreeClassifier(max_depth=2)
tree_classifier = tree_classifier.fit(X_train,y_train)
```

```python
##Again plotting the tree

plt.figure(figsize=(50,50))
tree.plot_tree(tree_classifier,filled = True)
```

```python
## Prediction

y_pred = tree_classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score , classification_report
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test, y_pred))
```
